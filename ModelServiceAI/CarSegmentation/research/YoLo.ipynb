```python
from fsspec.compression import unzip
from triton.language import argmax
from ultralytics import YOLO
import cv2
```

```python
model = YOLO("../../PeopleFlow/data/model_pt/yolo26n.pt")
```

```python
frame = cv2.imread("../data/example/img.png")
```

```python
res = model.predict(frame, classes=[2, 5, 7])
```

```python
import easyocr
reader = easyocr.Reader(['ru', 'en'])

result = reader.readtext(frame)

for (bbox, text, prob) in result:
    print(f"Номер: {text} (уверенность: {prob:.2f})")
```

```python
if len(res[0].boxes) > 0:
    best_box = max(res[0].boxes, key=lambda b: (b.xyxy[0][2] - b.xyxy[0][0]) * (b.xyxy[0][3] - b.xyxy[0][1]))
    x1, y1, x2, y2 = map(int, best_box.xyxy[0])
    crop = frame[y1:y2, x1:x2]

result = reader.readtext(crop)

for (bbox, text, prob) in result:
    print(f"Номер: {text} (уверенность: {prob:.2f})")
```
